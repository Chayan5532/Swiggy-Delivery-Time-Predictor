# Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

# Step 2: Load Dataset

```python
df = pd.read_csv(r"C:\Users\Aadesh Yogi\Desktop\ML Project\Swiggy Delivery Time Prediction\swiggy_demographic.csv")
```

```python
df
```

# Step 3: Data Preprocessing

```python
# Data types and non-null values
df.info()
```

# Step 4: Checking missing values

```python
df.isnull().sum()
```

# Step 5: Drop Irrelevant Columns

```python
df.columns
```

```python
cols_to_drop = [
    'rider_id',
    'order_date',
    'restaurant_latitude',
    'restaurant_longitude',
    'delivery_latitude',
    'delivery_longitude',
    'order_day',
    'order_month'
]

df = df.drop(columns=cols_to_drop, errors='ignore')
```

# Step 6: Feature Segregation

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
Nominal_cols = [
    'weather',
    'festival',
    'city_type',
    'type_of_order',
    'type_of_vehicle',
    'order_time_of_day',
    'order_day_of_week',
    'city_name'
]

Ordinal_cols = [
    'traffic'
]

Numerical_cols = [
    'age',
    'ratings',
    'vehicle_condition',
    'multiple_deliveries',
    'pickup_time_minutes',
    'order_time_hour',
    'distance',
    'is_weekend',
]
```

# Step 7: Build Preprocessing Pipelines

```python
# Nominal Pipeline (Encoding)

nominal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('One Hot Encoding', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

nominal_pipeline
```

```python
ordinal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal_encoding', OrdinalEncoder(
        categories=[['low', 'medium', 'high', 'jam']]
    ))
])

ordinal_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer
```

```python
# Numerical Pipeline (Scaling + Transformation)

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('Feature Scaling', StandardScaler()),
    ('Feature Transformation', PowerTransformer(method='yeo-johnson'))
])

numerical_pipeline
```

# Step 8: Apply ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
```

```python
Data_preprocessing = ColumnTransformer([('Nominal Transformation',nominal_pipeline,Nominal_cols),
                                        ('Ordinal Transformation', ordinal_pipeline, Ordinal_cols),
                                        ('Numerical Transformation',numerical_pipeline,Numerical_cols)])
Data_preprocessing
```


# Step 9: Define Features and Target

```python
X = df.drop('time_taken', axis=1)
y = df['time_taken']
```

# Step 10: Train-Test Split

```python
# Seggregate the training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

# Step 11: Train Machine Learning Models

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

## Step 12: Evaluate Model (R2)

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np

lr_model = Pipeline(steps=[
    ('preprocessing', Data_preprocessing),
    ('model', LinearRegression())
])

lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Linear Regression")
print("R2 Score:", r2)
print("MAE:", mae)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = Pipeline(steps=[
    ('preprocessing', Data_preprocessing),
    ('model', DecisionTreeRegressor())
])

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Decision Tree")
print("R2 Score:", r2)
print("MAE:", mae)
```

```python
from sklearn.svm import SVR

svr_model = Pipeline(steps=[
    ('preprocessing', Data_preprocessing),
    ('model', SVR())
])

svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("SVR")
print("R2 Score:", r2)
print("MAE:", mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = Pipeline(steps=[
    ('preprocessing', Data_preprocessing),
    ('model', RandomForestRegressor())
])

rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Random Forest")
print("R2 Score:", r2)
print("MAE:", mae)
```

```python
from xgboost import XGBRegressor

xgb_model = Pipeline([
    ('preprocessing', Data_preprocessing),
    ('model', XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        random_state=23
    ))
])

# Train
xgb_model.fit(X_train, y_train)

# Predict
y_pred = xgb_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("XGBoost")
print("R2 Score:", r2)
print("MAE:", mae)
```

## Overfitting Check

```python
# Train score
train_pred = xgb_model.predict(X_train)
print("Train R2:", r2_score(y_train, train_pred))

# Test score
test_pred = xgb_model.predict(X_test)
print("Test R2:", r2_score(y_test, test_pred))
```

# Step 13: Cross Validation

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    xgb_model,
    X,
    y,
    cv=5,
    scoring='r2'
)

print("Fold Scores:", scores)
print("Average R2:", scores.mean())
```

# Step 14: Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV

xgb_pipeline = Pipeline([
    ('preprocessing', Data_preprocessing),
    ('model', XGBRegressor(random_state=23))
])
```

```python
param_dist = {
    'model__n_estimators': [200, 300, 400, 500],
    'model__learning_rate': [0.01, 0.03, 0.05, 0.1],
    'model__max_depth': [4, 5, 6, 7, 8],
    'model__subsample': [0.7, 0.8, 1.0],
    'model__colsample_bytree': [0.7, 0.8, 1.0],
    'model__gamma': [0, 0.1, 0.2],
    'model__min_child_weight': [1, 3, 5]
}
```

```python
random_search = RandomizedSearchCV(
    xgb_pipeline,
    param_distributions=param_dist,
    n_iter=20,              # increase = better results
    cv=5,
    scoring='r2',
    verbose=2,
    n_jobs=-1,
    random_state=23
)

random_search.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = random_search.predict(X_test)

print("Final R2 Score:", r2_score(y_test, y_pred))
```

# Step 15: Select Best Model

```python
model = random_search
```

## Step 16: Prepare Input

```python
def prepare_input(distance, traffic, pickup_time, ratings, multiple_deliveries,
                  weather, festival, vehicle_condition):
    
    return pd.DataFrame([{
        'distance': distance,
        'traffic': traffic,
        'pickup_time_minutes': pickup_time,
        'ratings': ratings,
        'vehicle_condition': vehicle_condition,
        'multiple_deliveries': multiple_deliveries,
        'weather': weather,
        'festival': festival,

        # default / hidden features (model requirement)
        'age': 30,
        'city_type': 'urban',
        'city_name': 'INDO',
        'order_time_hour': 20,
        'type_of_order': 'meal',
        'type_of_vehicle': 'bike',
        'order_time_of_day': 'evening',
        'order_day_of_week': 'friday',
        'is_weekend': 0
    }])
```

# Step 17: Predict Output

```python
input_df = prepare_input(
    distance=5,
    traffic='high',
    pickup_time=15,
    ratings=4.2,
    multiple_deliveries=1,
    weather='sunny',
    festival='no',
    vehicle_condition=2
)

# Fix column order (VERY IMPORTANT)
input_df = input_df[X_train.columns]

prediction = model.predict(input_df)

print("Estimated Delivery Time:", round(prediction[0], 2), "minutes")
```

# Step 18: Overfitting Check

```python
# Train score
train_pred = model.predict(X_train)
print("Train R2:", r2_score(y_train, train_pred))

# Test score
test_pred = model.predict(X_test)
print("Test R2:", r2_score(y_test, test_pred))
```

# Step 19: Calculate MAE

```python
from sklearn.metrics import mean_absolute_error

y_pred_test = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_test)

print("Average Error (MAE):", round(mae, 2), "minutes")
```

# Step 20: Prediction Range

```python
pred = prediction[0]   # tumhara predicted value

lower = pred - mae
upper = pred + mae

print(f"Estimated Delivery Time: {pred:.2f} minutes")
print(f"Expected Range: {lower:.2f} to {upper:.2f} minutes")
```

# Step 21: Model Confidence

```python
confidence = 1 - (mae / y_test.mean())

print("Model Confidence:", round(confidence*100, 2), "%")
```
